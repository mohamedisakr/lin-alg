```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
v1 = np.array([4, 8, 10])
v2 = np.array([9, 2, 7])
```

```python
# many ways to calculate dot product

# first method
first = sum(np.multiply(v1, v2))
print(first )

# second method <recommended>
second = np.dot(v1, v2)
print(second)

# third method
third = np.matmul(v1, v2)
print(third)

# fourth method (using loop) just to show the mechanism of dot product
fourth = 0
for i in range(len(v1)):
    fourth += v1[i] * v2[i]

print(fourth)
```

#### distributive property

```python
# 3 vector
n = 3
a = np.random.randint(n)
b = np.random.randint(n)
c = np.random.randint(n)
```

```python
# a . (b + c) = a . b + a . c
ans1 = np.dot(a, (b+c))
ans2 = np.dot(a, b) + np.dot(a,c)

print(ans1)
print(ans2)
```

#### associative property

```python
# assoviative property hold in 2 cases
# 1. one of vectors is zero vector
# 2. a = b = c


# 3 vectors
n = 3
a = np.random.randint(n)
b = np.random.randint(n)
c = np.random.randint(n)
```

```python
# (a . b) . c = a . (b . c)
ans1 = np.dot(np.dot(a, b), c)
ans2 = np.dot(a, np.dot(b, c))

print(ans1)
print(ans2)
```

#### commutative property

```python
# 2 vectors
n = 3
a = np.random.randint(n)
b = np.random.randint(n)
```

```python
# ab , ba
ab = np.dot(a, b)
ba = np.dot(b, a)

print(ab)
print(ba)
```

#### dot product (geometrically)

```python
v1 = np.array([4, 8, 10])
v2 = np.array([9, 2, 7])
```

```python
v_dot = np.dot(v1, v2)
v_len = np.linalg.norm(v1) * np.linalg.norm(v2)
angle = np.arccos(v_dot / v_len)
angle
```

```python
# plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')

plt.axis((-6, 6, -6, 6)) 
plt.title(f'Angle between vectors: {angle} rad.') 
plt.show() 
```

```python
# compare algebraic & geometric dot product
algebraic = np.dot(v1, v2)
geometric = np.linalg.norm(v1) * np.linalg.norm(v2) * np.cos(angle)

print(f'algebraic result: {algebraic}')
print(f'geometric result: {geometric}')
```
