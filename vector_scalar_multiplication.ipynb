```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
# vector scalar multiplication
v = np.array([2,3])
s = 2
scaled_v = s * v
scaled_v
```

```python
# plot
plt.plot([0, v[0]],[0, v[1]],'b', label ='v_1') 
plt.plot([0, scaled_v[0]],[0, scaled_v[1]],'r:', label = '\lambda v_1') 
plt.axis('square') 
axlim = max([abs(max(v)),abs(max(scaled_v))])*1.5 # dynamic axis Lim 
plt.axis((-axlim,axlim,-axlim,axlim)) 
plt.grid() 
plt.show() 
```
